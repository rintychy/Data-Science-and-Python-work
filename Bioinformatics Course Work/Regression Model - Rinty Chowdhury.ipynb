```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
```

### Read the data file

```python
data = pd.read_csv('./Regression-Data.tsv', sep='\t', header=0)
data.head()
```

### Multiple linear regression

```python
X = data[['Gene Profile Size','Taxon Profile Size']]
Y = data['Semantic similarity']
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 
print("R-squared: " + str(model.rsquared))
```

### Residual plots

```python
sns.residplot(data['Gene Profile Size'], data['Semantic similarity'], color='red', scatter_kws={'s':4})
plt.title('Gene & Semantic', fontsize=14)
plt.xlabel('Gene Profile Size', fontsize=14)
plt.ylabel('Semantic similarity', fontsize=14)
plt.grid(True)
plt.show()
 
sns.residplot(data['Taxon Profile Size'], data['Semantic similarity'], color='blue', scatter_kws={'s':4})
plt.title('Taxon & Semantic', fontsize=14)
plt.xlabel('Taxon Profile Size', fontsize=14)
plt.ylabel('Semantic similarity', fontsize=14)
plt.grid(True)
plt.show()
```

### The R-squared value of the linear regression is 0.23 which is very low. This is a weak linear relationship and only 23% variation in Semantic similarity is explained by Gene Profile Size and Taxon Profile Size.

### Both of the residual plot shows linear relationship. However, none of them can predict the semantic similarity value correctly. The first residual plot shows the relationship between Gene profile size and semantic similarity. It has some outliers. All the points are far from the line, therefore, it's a weak relationship. The second residual plot shows the relationship between Taxon profile size and semantic similarity. All the points are also very far from the line. It has one outlier. However, there is a big gap of values between 1300 and 2700. Also as the Taxon size increases, the values are moving far from the line. First residual plot is better than the second plot because it doesn't contain any big gaps and the values are closer to the line compare to the second plot. Based on residual plots, Gene profile size can predict the semantic similarity better than Taxon profile size.
