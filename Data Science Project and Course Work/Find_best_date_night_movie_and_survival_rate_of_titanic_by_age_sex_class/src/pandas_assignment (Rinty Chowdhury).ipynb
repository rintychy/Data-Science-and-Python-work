##### Read in the movie data: `pd.read_table`

```python
def get_movie_data():
    
    unames = ['user_id','gender','age','occupation','zip']
    users = pd.read_table(os.path.join('../data','users.dat'), 
                          sep='::', header=None, names=unames)
    
    rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
    ratings = pd.read_table(os.path.join('../data', 'ratings.dat'), 
                            sep='::', header=None, names=rnames)
    
    mnames = ['movie_id', 'title','genres']
    movies = pd.read_table(os.path.join('../data', 'movies.dat'), 
                           sep='::', header=None, names=mnames)

    return users, ratings, movies
```

```python
users, ratings, movies = get_movie_data()
```

```python
print (users.head())
```

```python
print (ratings.head())
```

```python
print (movies.head())
```

##### Clean up the `movies`

- Get the `year`
- Shorten the `title`


```python
tmp = movies.title.str.extract('(.*) \(([0-9]+)\)')
tmp.apply(lambda x:x[0] if len(x) > 0 else None)
tmp.apply(lambda x: x[0][:40] if len(x) > 0 else None)
```

```python
movies['year'] = tmp[1]
movies['short_title'] = tmp[0]
```

```python
print (movies.head())
```

##### Join the tables with `pd.merge` (10 pts)

```python
MergedData = pd.merge(pd.merge(users, ratings, on='user_id'), movies, on='movie_id')
MergedData.head()
```

*** 3. Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, class and sex. (20 pts)***

Creating a new column name 'age_category' and iterate through the dataframe to fill the column properly

```python
t_df["age_category"] = ""

for x in range(1, len(t_df)+1):
    if (t_df.age[x] < 14):
        t_df.age_category[x] = "children"
    elif (t_df.age[x] >= 14) and (t_df.age[x] < 21):
        t_df.age_category[x] = "adolescents"
    elif (t_df.age[x] >= 21) and (t_df.age[x] < 65):
        t_df.age_category[x] = "adult"
    elif (t_df.age[x] >= 65):
        t_df.age_category[x] = "senior"
    else:
        t_df.age_category[x] = "NaN"
```

Creating seperate dataframes based on the age by using 'age-category' column
