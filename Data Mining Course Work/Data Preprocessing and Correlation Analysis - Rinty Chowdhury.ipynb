```python
import pandas as pd
import numpy as np
import scipy
import sklearn
from scipy import stats
from sklearn.preprocessing import MinMaxScaler
```

```python
my_data = pd.read_csv('../AssignmentMaterials/Video_Store.tsv', sep='\t', header=0)
my_data.head()
```

## 1. There are seven features in the data set.

## 2. There are four continuous features, and three nominal features.

## 3. Four continuous features are income, age, rentals, and avg per visit. Calculation of mean, median, maximum, minimum, and SD are below.

```python
# Removing "$", "," from Income column and coverting it from string to float
my_data["Income"] = my_data["Income"].str.replace('$','')
my_data["Income"] = my_data["Income"].str.replace(',','')
my_data['Income'] = my_data['Income'].astype(float)
```

```python
# Calculation for Income 
print("Mean income: " + str(my_data["Income"].mean()))
print("Median income: " + str(my_data["Income"].median()))
print("Maximum income: " + str(my_data["Income"].max()))
print("Minimum income: " + str(my_data["Income"].min()))
print("STD income: " + str(my_data["Income"].std()))
print()

# Calculation for Age 
print("Mean age: " + str(my_data["Age"].mean()))
print("Median age: " + str(my_data["Age"].median()))
print("Maximum age: " + str(my_data["Age"].max()))
print("Minimum age: " + str(my_data["Age"].min()))
print("STD age: " + str(my_data["Age"].std()))
print()

# Calculation for Rentals
print("Mean rentals: " + str(my_data["Rentals"].mean()))
print("Median rentals: " + str(my_data["Rentals"].median()))
print("Maximum rentals: " + str(my_data["Rentals"].max()))
print("Minimum rentals: " + str(my_data["Rentals"].min()))
print("STD rentals: " + str(my_data["Rentals"].std()))
print()

# Calculation for Avg_Per_Visit
print("Mean avg per visit: " + str(my_data["Avg Per Visit"].mean()))
print("Median avg per visit: " + str(my_data["Avg Per Visit"].median()))
print("Maximum avg per visit: " + str(my_data["Avg Per Visit"].max()))
print("Minimum avg per visit: " + str(my_data["Avg Per Visit"].min()))
print("STD avg per visit: " + str(my_data["Avg Per Visit"].std()))
```

## 4.1 Smoothing by bin means with bin depth of 4 of the 'Age' attribute

```python
# Sort the age
sorted_age = my_data["Age"].values
# Create numpy array using the sorted age
sorted_age = np.sort(sorted_age)
# Calculate the bin means, bin intervals, and bin labels using scipy.stats
mean_ages, bin_intervals, bin_labels = scipy.stats.binned_statistic(my_data.index, sorted_age, statistic='mean', bins=12)
# Remove 0 from the index and add the maximum value of the Age column into the intervals
bin_intervals = np.delete(bin_intervals, 0)
bin_intervals = np.append(bin_intervals, my_data["Age"].max())
# Create the new column with the calculated mean
my_data['Mean Age'] = pd.cut(x=my_data['Age'], bins=bin_intervals, labels=mean_ages)
```

## 4.2 Min-max normalization of the 'Income' attribute

```python
# Scale the data using MinMaxcScaler
scaler = MinMaxScaler()
# Reshape the data column in order to fit the scaler
data = my_data["Income"].values.reshape(-1,1)
# Fit the data into the scaler and scale the value
scaler.fit(data)
# Create new column with the scaled value
my_data['Normalized Income'] = scaler.transform(data)
```

## 4.3 Z-score normalization of the 'Rentals' attribute

```python
# Normalizing the rentals value using z-score and creating new column with that value
my_data['Normalized Rentals'] = scipy.stats.mstats.zscore(my_data["Rentals"])
```

## 4.4 Discretize the original 'Income' attribute

```python
# Create new column based on the given bin intervals and labels
my_data["Income Category"] = pd.cut(x=my_data['Income'], bins=[0, 25000, 59000, my_data["Income"].max()], labels=["Low", "Mid", "High"])
my_data.head(50)
```

## 5. Converting categorical attribute into multiple attributes
